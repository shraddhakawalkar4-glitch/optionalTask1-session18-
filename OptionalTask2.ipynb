```python

plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.countplot(x="work_year",data=df)
plt.title("Distribution of work_year")
plt.xlabel("Work year")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.histplot(df["salary"],bins=30)
plt.title("Distribution of salary")
plt.xlabel("salary")
plt.ylabel("Count")

plt.subplot(2,2,3)
sns.histplot(df["salary_in_usd"])
plt.title("Distribution of salary_in_usd")
plt.xlabel("salary_in_usd")
plt.ylabel("Count")

plt.subplot(2,2,4)
sns.histplot(df["remote_ratio"])
plt.title("Distribution of remote_ratio")
plt.xlabel("remote_ratio")
plt.ylabel("Count")

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['salary'], df['salary_in_usd'])
plt.title("Salary vs Salary in USD")
plt.xlabel("Salary")
plt.ylabel("Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="company_size", y="salary_in_usd", data=df, palette="Set2")
plt.title("Salary Distribution by Company Size")
plt.xlabel("Company Size")
plt.ylabel("Salary in USD")
plt.show()
```

```python
sns.countplot(x="employment_type", data=df, palette="Set2")
plt.title("Distribution of Employment Types")
plt.xlabel("Employment Type")
plt.ylabel("Count")
plt.show()

sns.countplot(x="experience_level", data=df, palette="Set2")
plt.title("Distribution of Experience Levels")
plt.xlabel("Experience Level")
plt.ylabel("Count")
plt.show()

sns.countplot(x="company_size", data=df, palette="Set2")
plt.title("Distribution of company_size")
plt.xlabel("company_size")
plt.ylabel("Count")
plt.show()
```

```python
# Countplot Insights:
# - Senior (SE) & Mid-level (MI) roles dominate
# - Full-time (FT) jobs are most common
# - Data Scientist, Data Engineer, Data Analyst appear most often
# - USD is the primary salary currency
# - US leads in employee residence & company location
# - Large (L) companies hire the most
```

```python
corr=df[['work_year','salary','salary_in_usd','remote_ratio']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

```

```python
# Insights:
# - salary is highly correlated with salary_in_usd (conversion)
# - work_year shows mild correlation with salary_in_usd (salary trends over years)
# - remote_ratio has little to no correlation with salary_in_usd
```

```python
# Count distribution
company_counts = df['company_size'].value_counts()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(company_counts, labels=company_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Company Size")
plt.show()
```
