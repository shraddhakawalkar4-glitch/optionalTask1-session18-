```python
df["expenses"].plot(kind="hist",edgecolor="black")
plt.title("Expenses Distribution")
plt.xlabel("Expenses")
plt.ylabel("Frequency")
plt.show()

df["bmi"].plot(kind="hist",edgecolor="black")
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()

df["age"].plot(kind="hist",edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
print("\nScatter plot:")
plt.figure(figsize=(8,6))
plt.scatter(df["expenses"],df["age"])
plt.title("Scatter Plot of Expenses vs Age")
plt.xlabel("Expenses")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="smoker",y="expenses",data=df,hue="region",color="red")
plt.xticks(rotation=45)
plt.title("Box Plot of Expenses by Smoker Status")
plt.xlabel("Smoker Status")
plt.ylabel("Expenses")
plt.show()
```

```python
print("\nCategorical Columns count plots")

sns.countplot(x="sex", data=df)
plt.title("sex  Count Plot")
plt.xlabel("sex")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

sns.countplot(x='smoker', data=df)
plt.title("Smoker  Count Plot")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

sns.countplot(x='region', data=df)
plt.title("Region Count Plot")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()

plt.tight_layout()
plt.show()
```

```python
# Insights from categorical count plots:

# 1. Sex:
# The dataset is fairly balanced between male and female.
# Insight: No strong gender bias in sample size, so comparisons across sex are reliable.

# 2. Smoker:
# Majority are non-smokers, but smokers form a significant minority.
# Market trend: Smoking status is a critical factor — smokers tend to have much higher expenses.

# 3. Region:
# All four regions (northeast, northwest, southeast, southwest) are represented.
# Distribution is fairly even, though southeast and southwest often have slightly more entries.
# Insight: Dataset is geographically diverse, so regional comparisons are possible.
```

```python
corr=df[['age','bmi','children','expenses']].corr()
sns.heatmap(corr,annot=True,cmap="magma")
plt.title("Correlation Heatmap")
```

```python
# Correlation with target (expenses):

# 1. Age vs Expenses:
# Positive correlation — older individuals generally have higher medical expenses.
# Insight: Age is a strong driver of costs, reflecting increased health risks with age.

# 2. BMI vs Expenses:
# Moderate positive correlation — higher BMI often leads to higher expenses.
# Insight: Obesity-related health issues contribute to increased insurance charges.

# 3. Children vs Expenses:
# Weak correlation — number of children has little direct effect on expenses.
# Insight: Family size doesn’t strongly influence costs compared to lifestyle/health factors.

# 4. Smoker (categorical, but impactful):
# Strong effect — smokers consistently show much higher expenses than non-smokers.
# Insight: Smoking status is the single most important categorical factor influencing expenses.

# 5. Region:
# Very weak correlation — expenses are fairly evenly distributed across regions.
# Insight: Geography doesn’t play a major role in cost variation.
```

```python
# Count values in 'region'
region_counts = df['region'].value_counts()

# Create pie chart
plt.figure(figsize=(7,7))
plt.pie(
    region_counts,
    labels=region_counts.index,
    autopct='%1.1f%%',
    startangle=90,
)
plt.title("Distribution of Insurance Records by Region")
plt.show()
```
