```python
print("\n---HISTOGRAM---")
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.hist(df['revenue'],edgecolor="black")
plt.title("Distribution of Revenue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")

plt.subplot(2,2,2)
plt.hist(df['quantity'],edgecolor="black")
plt.title("Distribution of Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")

plt.subplot(2,2,3)
plt.hist(df['unit_price'],edgecolor="black")
plt.title("Distribution of Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")

plt.subplot(2,2,4)
plt.hist(df["discount"],edgecolor="black")
plt.title("Distribution of Discount")
plt.xlabel("Discount")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()


```

```python
plt.figure(figsize=(10,8))
sns.regplot(x='unit_price',y='revenue',data=df,line_kws={'color':'red'})
plt.title("Scatter Plot of unit_price Vs 'revenue")
plt.xlabel("Unit price")
plt.ylabel("'revenue")
# plt.scatter(df['year'],df['revenue'])
# plt.title("Scatter Plot of Year vs Revenue")
# plt.xlabel("year")
# plt.ylabel("revenue")
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='product_category',y='revenue',data=df,hue='product_category',palette='Set2')
plt.xticks(rotation=45)
plt.title("Box Plot of product_category vs Revenue")
plt.xlabel("product_category")
plt.ylabel("revenue")
```

```python
print("\nCategorical count plots")
plt.figure(figsize=(10,6))
sns.countplot(x="product_category",data=df,palette="Set2",hue="region")
plt.title("Count Plot of product_category")
plt.xlabel("product_category")
plt.ylabel("Count")
plt.show()

sns.countplot(x="payment_method",data=df,palette="Set2",hue="region")
plt.title("Count Plot of payment_method")
plt.xlabel("payment_method")
plt.ylabel("Count")
plt.show()

sns.countplot(x="region",data=df,palette="Set2")
plt.title("Count Plot of region")
plt.xlabel("region")
plt.ylabel("Count")
plt.show()



plt.tight_layout()
plt.show()
```

```python
# Key categorical insights:
# 1. Region: Orders are evenly distributed across North, South, East, and West,
#    showing balanced geographic coverage without dominance by any single region.
# 2. Product Category: Clothing and Electronics have noticeably higher order counts,
#    making them the core revenue drivers compared to Beauty and Home.
# 3. Payment Method: Card is the most frequently used payment option, followed by COD,
#    while Wallet is less common but still relevant in certain regions.
# 4. Customer Rating: Most ratings cluster around 3–4 (Medium to High),
#    with fewer extreme lows or perfect 5s, suggesting moderate satisfaction overall.
```

```python
corr=df[['quantity','unit_price','revenue','customer_rating','delivery_days','discount']].corr()
sns.heatmap(corr,annot=True)
plt.title("Correlation Heatmap")
```

```python
# Correlation insights with target variable (revenue):
# 1. unit_price → Strong positive correlation with revenue.
#    Higher unit prices generally lead to higher order revenue.
# 2. quantity → Positive correlation with revenue.
#    Larger order sizes contribute directly to higher revenue.
# 3. discount → Mild negative correlation with revenue.
#    Higher discounts reduce effective revenue per order.
# 4. delivery_days → Weak or negligible correlation with revenue.
#    Delivery speed does not strongly affect revenue generation.
# 5. customer_rating → Weak correlation with revenue.
#    Ratings reflect satisfaction but do not directly drive revenue.
```

```python
region_count=df['payment_method'].value_counts()
plt.figure(figsize=(10,6))
plt.pie(region_count,labels=region_count.index,autopct='%1.1f%%',startangle=90)
plt.title("Distribution of Payment method")
plt.show()
#
```
